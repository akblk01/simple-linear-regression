```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataFrame = pd.read_csv('50_Startups.csv')
```

```python
dataFrame.head(1)
```

```python
dataFrame = dataFrame.drop('State', axis = 1)
```

```python
dataFrame.head(3)
```

```python
import seaborn as sns
sns.jointplot(x = 'Marketing Spend', y = 'Profit', data = dataFrame, kind = 'reg')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x = dataFrame['Marketing Spend']
y = dataFrame['Profit']
x = x.to_numpy().reshape(-1,1)
y = y.to_numpy().reshape(-1,1)
```

```python
reg = LinearRegression()
```

```python
model = reg.fit(x, y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
model.score(x,y) ## Market harcamalarının kara etkisi
```

```python
model.predict([[471784.1]])
```

```python
realPredict = pd.DataFrame(y[0:10])
```

```python
guessPredict = pd.DataFrame(model.predict(x)[0:10])
```

```python
hatalar = pd.concat([realPredict, guessPredict], axis = 1)
```

```python
hatalar.columns = ['Reel_Predict', 'Guess_Predict']
```
